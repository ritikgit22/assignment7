Q5 - Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 100
n_features = 5

# Create a correlation matrix
corr_matrix = np.array([[1.0, 0.8, 0.6, 0.4, 0.2],
                        [0.8, 1.0, 0.7, 0.5, 0.3],
                        [0.6, 0.7, 1.0, 0.6, 0.4],
                        [0.4, 0.5, 0.6, 1.0, 0.5],
                        [0.2, 0.3, 0.4, 0.5, 1.0]])

# Generate the dataset based on the correlation matrix
data = np.random.multivariate_normal(mean=[0] * n_features, cov=corr_matrix, size=n_samples)

# Create a DataFrame from the data
df = pd.DataFrame(data, columns=[f'Feature {i+1}' for i in range(n_features)])

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid for x and y
x, y = np.meshgrid(x, y)

# Calculate z values using the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame for the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])
fig.update_layout(
    title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Output the plot in the notebook
output_notebook()

# Create a figure object
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Generate x and y values
x = np.linspace(0, 10, 200)
y = np.sin(x)

# Plot the line
p.line(x, y, line_width=2, line_color="blue")

# Show the plot
show(p)
```

Q2 - Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

Q3 - Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

    fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

    counts = [20, 25, 30, 35]

```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.transform import dodge
from bokeh.models import ColumnDataSource

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Creating the data source
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Creating a figure
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Adding bars to the figure
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="skyblue")

# Customizing the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Counts"
p.xaxis.axis_label = "Fruits"

# Display the plot
show(p)
```

Q4 - Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.io import show
from bokeh.plotting import figure

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Random Data", background_fill_color="#fafafa")

# Add a quad glyph for the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Customizing the plot
p.yaxis.axis_label = "Frequency"
p.xaxis.axis_label = "Data Values"
p.grid.grid_line_color="white"

# Display the plot
show(p)
```
